```python
import matplotlib.pyplot as plt
import numpy as np
import time


def linear_search(arr,item):
    for x in range(len(arr)):
        if(arr[x]==item):
            return

# Generate input sizes (n)
input_sizes = np.arange(1000, 10001, 100)
# Perform time measurements
time_measurements = []
for size in input_sizes:
    dataset = list(range(1, size + 1)) 
    start_time = time.time()
    linear_search(dataset,size-1)
    end_time = time.time()
    time_taken = end_time - start_time
    time_measurements.append(time_taken)
time_measurementbs = []
for size in input_sizes:
    dataset = list(range(1, size + 1)) 
    start_time = time.time()
    if(size in dataset):
        pass
    end_time = time.time()
    time_taken = end_time - start_time
    time_measurementbs.append(time_taken)




```

```python
# Plot the time complexity
plt.plot(input_sizes, time_measurements)
plt.plot(input_sizes, time_measurementbs)
plt.xlabel('Input Size (n)')
plt.ylabel('Time Taken (seconds)')
plt.title('So')
plt.show()
```
